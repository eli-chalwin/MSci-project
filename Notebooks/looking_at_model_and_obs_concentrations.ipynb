**<h2>Looking at modeled and observed ch4 concentrations for different DECC sites</h2>** 

First import required libraries:

```python
from openghg.retrieve import  search
import functions as f 
```

Then look at the sites in the DECC network:

```python
search(network='decc',species='ch4',data_type='surface',store='obs_paris_2024_07_store').results #search for avaliable sites 
```

Next select a site and inlet to look at:

```python
#define site and inlet 
site = "bsd"
inlet = "248"
```

Next view the observational data for this site and inlet:

```python
obs = f.obs(site, inlet) #function that retrieves and plots observational data - see functions file for code 
f.plot(obs, "Monthly average of observed CH4 for bsd248", "Time", "ch4 conc. (ppb)") #function that plots a graph
```

We can then use openghg to construct a model to predict ch4 concentrations making uses of flux, boundary condition and footprint data:

```python
model = f.scenario(site, inlet, "model") #function that sets up a modelscenario 
f.plot(model, "Modelled monthly average ch4 conc for bsd248", "Time", "ch4 conc. (ppb)") #function that plots a graph
```

There are a two important things to notice from this first model. First is that the modelled ch4 concentrations are far below the observational data. This is because the model doesn't account for the background ch4 concentration. To fix this we can simply calculate a baseline and add it to the model.

The second thing to notice is that the model fails prior to 2015. This is because the flux (emissions) data only covers 2015-2022, as such the model can only predict within this time range. This can be fixed if desired by taking a timeslice of the data set.

Now that we've calculated a simple model, lets calculate and plot a baseline:

```python
baseline = f.scenario(site, inlet, "baseline") #function that sets up an openghg modelscenario and adds the relevant data and then constructs a baseline - see functions file for code
f.plot(baseline, "baseline ch4 conc for bsd248", "Time", "ch4 conc. (ppb)") #function that plots a graph
```

Here we can see the baseline ch4 concentration and how it varies seasonally. However, again notice that the baseline model fails prior to 2018. This is because the boundary condition data we're using, "camv22r2_daily", only covers 2018-01-01 to 2023-12-31. Another boundary condition file called "camsv19_daily" exists which covers 2000-01-01 to 2019-12-31 which we could also use.

For simplicity, in this notebook we are going to find the overlap between the flux, boundary condition and footprint data and then constrain the "Time" variable for each to be within the overlap.

Find the overlap:

```python
f.overlap(site, inlet) #function that finds the time period overlap between boundary condition, flux and footprint data
```

Now that we have the overlap, we can calculate the model and the baseline between these two dates and then add them together to get a final model to compare to the observational data.

```python
start_date, end_date = f.overlap(site, inlet) #sets start and end dates
model = f.scenario(site, inlet, "model", start_date, end_date ) #construct model
baseline = f.scenario(site, inlet, "baseline", start_date, end_date ) #construct baseline
combined = model + baseline #add the two together   
f.plot(baseline, "Modelled monthly average baseline for bsd248", "Time", "ch4 conc. (ppb)") #plot combined model
f.plot(model, "Modelled monthly average ch4 conc exc.baseline for bsd248", "Time", "ch4 conc. (ppb)") #plot combined model
f.plot(combined, "Modelled monthly average ch4 conc inc.baseline for bsd248", "Time", "ch4 conc. (ppb)") #plot combined model #if you exclude obs data from model scenario then combined model looks different? 
```

We can also replot the observed data between these two dates to make the two graphs easier to compare.

```python
obs = f.obs(site, inlet, start_date, end_date) #function that retrieves and plots observational data - see functions file for code 
f.plot(obs, "Monthly average of observed CH4 for bsd248", "Time", "ch4 conc. (ppb)") #function that plots a model 
```

Here we can see in this instance the model shows similar trends to the observed data. However, it seems that the model tends to underpredict ch4 conc. when compared to the observational data. (This may be caused due to the baseline under predicting (?).)

Alternatively we can compare the two in a more objective manner by using some simple stats

```python
std, rmse, corr = f.stats(obs, combined) #function that calculates std, rmse and corr
print(f"RMSE: {rmse}\n"  f"Standard Deviation: {std}\n" f"Pearson correlation coefficient: {corr}")
```

We can then plot these stats on a Taylor graph. Although this doesn't serve much purpose here, Taylor graphs can be used to easily summarise the stats for several different models.

```python
f.taylor(std, rmse, corr, 'bsd248') #function to plot a taylor diagram 
```
