**<h2>Summarising stats for different DECC sites using Taylor diagrams</h2>** 

```python
import functions as f 
```

First let us construct models for the sites whose stats we want to look at. Here we are only looking at three different sites. This is because these are the only sites where the footprint and flux data overlap with the "camv22r2_daily" boundary conditions. If we were to construct models using the "camsv19_daily" boundary conditions there would likely be more sites with an overlap.

```python
bsd248_model = f.scenario("bsd", "248", "combined", '2018-01-01', '2021-08-31' ) 
bsd248_obs = f.obs("bsd", "248", '2018-01-01', '2021-08-31')
rgl90_model = f.scenario("rgl", "90", "combined", '2018-01-01', '2022-12-31' ) 
rgl90_obs = f.obs("rgl", "90", '2018-01-01', '2022-12-31')
hfd100_model = f.scenario("hfd", "100", "combined", '2018-01-01', '2022-12-31' ) 
hfd100_obs = f.obs("hfd", "100", '2018-01-01', '2022-12-31')
```

Next lets calculate the stats for each of these sites:

```python
bsd248_std, bsd248_rmse, bsd248_corr = f.stats(bsd248_obs, bsd248_model)
rgl90_std, rgl90_rmse, rgl90_corr = f.stats(rgl90_obs, rgl90_model)
hfd100_std, hfd100_rmse, hfd100_corr = f.stats(hfd100_obs, hfd100_model)
```

Now lets make a list for each of the stats to feed into the Taylor diagram function:

```python
std = [bsd248_std, rgl90_std, hfd100_std]
rmse = [bsd248_rmse, rgl90_rmse, hfd100_rmse]
corr = [bsd248_corr, rgl90_corr, hfd100_corr]
sites = ['bsd248', 'rgl90', 'hfd100']
```

Finally lets plot a Taylor diagram to summarise the stats for each site:

```python
f.taylor(std, rmse, corr, sites) #function to plot a taylor diagram 
```
