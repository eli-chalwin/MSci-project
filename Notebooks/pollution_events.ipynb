**<h2>Looking at pollution events for different DECC sites</h2>** 

Instead of combining modelled emissions with the calculated baseline, we may instead want to look at pollution events by removing the baseline from the observational data.

```python
import functions as f 
#define site and inlet 
site = "bsd"
inlet = "248"
f.overlap(site, inlet) #function that finds the time period overlap between boundary condition, flux and footprint data
start_date, end_date = f.overlap(site, inlet) #sets start and end dates
obs = f.obs(site, inlet, start_date, end_date) #function that retrieves and plots observational data - see functions file for code 
baseline = f.scenario(site, inlet, "baseline", start_date, end_date) #function that sets up an openghg modelscenario and adds the relevant data and then constructs a baseline - see functions file for code
pollution_events = obs - baseline 
f.plot(pollution_events, "pollution events for bsd248", "Time", "ch4 conc. (ppb)") #function that plots a graph
```

We can then compare this to the model to see how well our model predicts these pollution events.

```python
model = f.scenario(site, inlet, "model", start_date, end_date ) #construct model
f.plot(model, "Modelled monthly average ch4 conc exc.baseline for bsd248", "Time", "ch4 conc. (ppb)") #plot combined model
```

To compare the model and the observed pollution events (against the modelled baseline) we can use some simple stats:

```python
std, rmse, corr = f.stats(pollution_events, model) #function that calculates std, rmse and corr
print(f"RMSE: {rmse}\n"  f"Standard Deviation: {std}\n" f"Pearson correlation coefficient: {corr}")
```

Note that the rmse and std are quite high and the correlation relatively poor. This suggests that for bsd248 the modelled pollution events are not particularly accurate.
